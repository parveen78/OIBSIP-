The unemployment rate in India, amidst lockdown and restrictions on mobility, is 12.81% as of June 8th 2021 based on the data provided by the CMIE. Earlier, the unemployment rate in India shot up from 6.5 per cent in March 2021 to 8 per cent in April 2021, to 14.7% by May end, while the employment rate fell from 37.6 per cent in March to 36.8 per cent in April, says the report of CMIE – Centre For Monitoring Indian Economy.

In 2020, the unemployment rate in India fell to 7% in September 2020 from the record high of 29% since the country went into lockdown from March 2020,  However, it later increased to 9.1% in December 2020.

The unemployment rate again declined to 6.5 per cent in January 2021 from 9.1 per cent in December 2020, while the employment rate surged to 37.9 per cent as compared to 36.9 per cent.

The lockdown to contain the coronavirus outbreak has forced many industries to shut down thus increasing unemployment across the country.

#DATA VISULAIZATION

```python
plt.style.use("seaborn-whitegrid") # here we do correlation of unemployment dataset
plt.figure(figsize=(12,10))
sns.heatmap(Unemployment.corr())
plt.show()
```

NOW WE HAVE TO LOOK ESTIMATE NUMBER OF EMPLOYEES ACCORDING TO REGIONS

```python
Unemployment.columns=("Region","Date","Frequency",                           
 "Estimated Unemployment Rate (%)",           
 "Estimated Employed",                        
 "Estimated Labour Participation Rate (%)",    
"Region.1",                                   
"longitude",                                   
"latitude ")
plt.title("Indian Employment")
sns.histplot(x='Estimated Employed',hue='Region.1',data=Unemployment)
plt.show()
```

#As it show in this histogram bar chart the maximum number of employed in north and northeast region.

```python
Unemployment.columns=("Region","Date","Frequency",                           
 "Estimated Unemployment Rate (%)",           
 "Estimated Employed",                        
 "Estimated Labour Participation Rate (%)",    
"Region.1",                                   
"longitude",                                   
"latitude ")
fig = px.bar(Unemployment, x="Date", y="Estimated Unemployment Rate (%)", color='Region.1', 
              title='Estimated unemployement rate according to date', width=1000, height=500,
             labels={"Estimated Unemployment Rate (%)": "Estimated Unemployment Rate ", "Date": "Date"})
fig.show()
```

#So it is clearly from this bar chart that mostly people from north region are unemployed and maximun in the month of may with the 42.27% unemployment rate.

#Comparision between indian employment and unemployement according to region

```python
Unemployment.columns=("Region","Date","Frequency",                           
 "Estimated Unemployment Rate (%)",           
 "Estimated Employed",                        
 "Estimated Labour Participation Rate (%)",    
"Region.1",                                   
"longitude",                                   
"latitude ")
fig = px.scatter(Unemployment, x="Estimated Unemployment Rate (%)", y="Estimated Employed", color="Region", width=800, height=400,
            labels={"Estimated Unemployment Rate (%)": "Estimated Unemployment Rate in year 2020", "Estimated Employed": "Estimated Employed in year 2020"})
fig.show()
```

# So in this visulaization the people from north region mostly the state from uttarpradhesh 7.58% have high unemployment rate.

```python
Unemployment.columns=("Region","Date","Frequency",                           
 "Estimated Unemployment Rate (%)",           
 "Estimated Employed",                        
 "Estimated Labour Participation Rate (%)",    
"Region.1",                                   
"longitude",                                   
"latitude ")
fig = px.scatter(Unemployment,x="Estimated Unemployment Rate (%)", y="Estimated Employed" , facet_col="Region.1", color="Region",
            labels={"Estimated Unemployment Rate (%)": "Estimated Unemployment Rate in year 2020", "Estimated Employed": "Estimated Employed in year 2020"})
fig.show()
```
